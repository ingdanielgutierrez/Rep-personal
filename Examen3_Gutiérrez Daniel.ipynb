#### Ejercicio 2.  Ecuaciones de Lorenz
Un modelo simple basado en la dinámica de fluidos atmosféricos es representado mediante las ecuaciones de Lorenz, desarrollado por el meteorólogo Estadounidense Edward Lorenz:
\begin{eqnarray}
\frac{dx}{dt}&=&-\sigma x+\sigma y\\
\frac{dy}{dt}&=&rx-y-xz\\
\frac{dz}{dt}&=&-bz+xy
\end{eqnarray}
Lorenz desarrolló estas ecuaciones para representar la intensidad de movimiento de un fluido atmosférico. Donde $x$ se refiere a la variación de temperatura, $y$ y $z$ a las direcciones en el plano horizontal y vertical respectivamente.

- 2.1 Defina la función que describa el modelo completo. Los argumentos deberán ser:


1. El arreglo $[x,y,z]$,
2. La variable de tiempo $t$,
3. El parámetro $\sigma$,
4. El parámetro $r$,
5. El parámetro $b$.

    Y la salida que sea el arreglo de derivadas $\left[\frac{dx}{dt},\frac{dy}{dt},\frac{dz}{dt}\right]$.

```python
def fun(t, a):
    x = a[0]
    y = a[1]
    z = a[2]
    dx = -s*x + s*y
    dy = r*x - y - x*z
    dz = -b*z + x*y
    return np.array([dx, dy, dz])
```

- 2.2 Usando $\sigma=10$, $b=\frac{8}{3}$ y $r=28$; condiciones iniciales $x_0=y_0=z_0=5$; encuentre la solución numérica y grafique la variación de temperatura con respecta al tiempo, de 0 a 20 segundos.

```python
#condiciones iniciales
s = 10
b = (8/3)
r = 28
x0 = np.array([5, 5, 5])

#vector de tiempo
tspan = [0, 20]

#resolver numéricamente
sol = solve_ivp(fun, tspan, x0)

#separar soluciones
x = sol.y[0]
y = sol.y[1]
z = sol.y[2]

#gráfica de x vs t
plt.figure()
plt.plot(sol.t, x)
plt.xlabel('Tiempo')
plt.ylabel('Var Temperatura')
plt.grid()
```

- 2.3 Grafique el plano de fase de $x$ vs $y$.

```python
#gráfica de x vs y
plt.figure()
plt.plot(y, x)
plt.xlabel('Var Temperatura')
plt.ylabel('Dir Horizontal')
plt.grid()
```

#### Ejercicio 3. Oscilador armónico
Un sistema simple de oscilación armónica es el péndulo descrito por la siguiente ecuación:
<div>
<img style="float: left; margin: 0px 0px 15px 15px;" src="https://upload.wikimedia.org/wikipedia/commons/7/76/Pendulum.jpg" width="150px" height="50px" />
</div>

$$
\frac{d^2 \theta}{dt^2} + \omega_{0}^{2}\, \theta = 0, \quad\mbox{donde}\quad \omega_{0}^2 = \frac{g}{l} 
$$

La solución está descrita por
\begin{eqnarray}
\theta(t) &=& \theta(0) \cos(\omega_{0} t) + \frac{\dot{\theta}(0)}{\omega_{0}} \sin(\omega_{0} t)\\
\dot{\theta}(t)&=&-\omega_0\theta(0)\sin(\omega_0t)+\dot{\theta}(0)\cos(\omega_0t).
\end{eqnarray}

- 3.1 Defina la función de la ecuación diferencial para el pédulo simple, cuyos argumentos de entrada sean:


1. el arreglo $[\theta,\dot{\theta}]$,
2. el tiempo $t$,
3. la frecuencia característica $\omega_0$.

  Considerando que la salida es el arreglo $[\dot{\theta},\ddot{\theta}]$.

```python
def eq(t, Th):
    th = Th[0]
    dth = Th[1]
    return np.asarray([dth, -w0**2*th])
```

- 3.2 Usando $l=1\,m$ y condiciones iniciales $[\theta(0),\dot{\theta}(0)]=[0.1,0]$. Encuentre la solución numérica para $\theta(t)$ y grafíque $\theta$ vs $t$ entre 0 y 2 segundos.

```python
#valor de w0
g = 9.81
l = 1
w0 = np.sqrt(g/l)

#condiciones iniciales
th0 = 0.1
dth0 = 0

tspan = [0, 2]

#solucion ecuacion
sol = solve_ivp(eq, tspan, [th0, dth0], max_step=0.1)
sol.y
```

```python
#gráfica de la solución numérica
plt.figure()
plt.plot(sol.t, sol.y[0, :], 'r', label='Solución numérica')
plt.xlabel('Tiempo')
plt.ylabel('Posición')
plt.legend(loc='best')
plt.grid()
```

- 3.3 Defina la función para la solución analítica y compare en una gráfica con la solución numérica

```python
def an(t, w0):
    C1 = th0
    C2 = dth0/w0
    return C1*np.cos(w0*t) + C2*np.sin(w0*t)
```

```python
plt.figure()
plt.plot(sol.t, sol.y[0, :], 'r', lw=5, label='Solución numérica')

t = np.linspace(0, 2, 1000)
posA = an(t, w0)
plt.plot(t, posA, 'b', lw=2, label='Solución análitica')
plt.xlabel('Tiempo')
plt.ylabel('Posición')
plt.legend(loc='best')
plt.grid()
```
